```python
import pandas as pd
df=pd.read_csv('/content/top_bots_clean.csv.zip')
df.head(10)
```

```python
df.info()
```

```python
taxonomy = {
    "Gaming": ["game", "valorant", "csgo", "pubg", "fps", "gamer"],
    "Anime": ["anime", "manga", "weeb", "otaku"],
    "Music": ["music", "spotify", "song", "playlist", "dj"],
    "Education": ["study", "student", "learn", "homework", "school"],
    "Technology": ["tech", "ai", "coding", "python", "bot", "developer"],
    "Social": ["chat", "friends", "hangout", "meet"],
}
```

```python
def tag_description(description):
    if pd.isna(description):
        return []
    tags = []
    for category, keywords in taxonomy.items():
        if any(keyword.lower() in description.lower() for keyword in keywords):
            tags.append(category)
    return tags
```

```python
df['Predicted Tags'] = df['long_description'].apply(tag_description)

```

```python
print(df.columns)
```

```python
from google.colab import files

df.to_csv("tagged_servers.csv", index=False)
files.download("/content/tagged_servers.csv") # Now 'files' is defined and can be used
```

```python
import matplotlib.pyplot as plt
from collections import Counter

all_tags = [tag for sublist in df['Predicted Tags'] for tag in sublist]
tag_counts = Counter(all_tags)

plt.figure(figsize=(10, 6))
plt.bar(tag_counts.keys(), tag_counts.values(), color='skyblue')
plt.title("Distribution of Predicted Tags")
plt.ylabel("Number of Servers")
plt.xlabel("Tag Category")
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```
