**20. Classify movies based on ratings[GOOD,BETTER and BEST].**

```python
imdb.columns
```

```python
def rating(rating):
  if rating>=8.0:
    return 'BEST'
  elif rating>=7.0:
    return 'GOOD'
  elif rating>=6.0:
    return 'BETTER'
  else:
    return 'AVERAGE'
```

```python
imdb['Rating_Category']=imdb['Rating'].apply(rating)
```
