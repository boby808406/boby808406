```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
from IPython.display import display
```

```python
df['Datetime'] = pd.to_datetime(df['Datetime'], format='%d-%m-%Y %H:%M')
df['Date'] = df['Datetime'].dt.date
df['Hour'] = df['Datetime'].dt.hour
df['Day'] = df['Datetime'].dt.day
df['Month'] = df['Datetime'].dt.month
df['Slot'] = np.where(df['Hour'] < 12, 'S1', 'S2')
df['Games_Played'] = 10
```

Part A - Calculating loyalty points

On each day, there are 2 slots for each of which the loyalty points are to be calculated:
S1 from 12am to 12pm
S2 from 12pm to 12am"
"Based on the above information and the data provided answer the following questions:
1. Find Playerwise Loyalty points earned by Players in the following slots:-
    a. 2nd October Slot S1
    b. 16th October Slot S2
    b. 18th October Slot S1
    b. 26th October Slot S2

```python
def calculate_loyalty_points(data, target_date, slot):

    filtered = data[(data['Date'] == target_date) & (data['Slot'] == slot)]
    result = filtered.groupby('User Id').agg(
        Total_Deposits=('Amount', lambda x: x[x > 0].sum()),
        Total_Withdrawals=('Amount', lambda x: abs(x[x < 0].sum())),
        Deposit_Count=('Amount', lambda x: (x > 0).sum()),
        Withdrawal_Count=('Amount', lambda x: (x < 0).sum())
    ).reset_index()

    result['Games_Played'] = 10
    result['Loyalty_Points'] = (
        0.01 * result['Total_Deposits'] +
        0.005 * result['Total_Withdrawals'] +
        0.001 * np.maximum(result['Deposit_Count'] - result['Withdrawal_Count'], 0) +
        0.2 * result['Games_Played']
    )

    return result.sort_values('Loyalty_Points', ascending=False)

print("\n📅 2nd October Slot S1 Results:")
oct2_s1 = calculate_loyalty_points(df, pd.to_datetime('2022-10-02').date(), 'S1')
display(oct2_s1)

print("\n📅 16th October Slot S2 Results:")
oct16_s2 = calculate_loyalty_points(df, pd.to_datetime('2022-10-16').date(), 'S2')
display(oct16_s2)

print("\n📅 18th October Slot S1 Results:")
oct18_s1 = calculate_loyalty_points(df, pd.to_datetime('2022-10-18').date(), 'S1')
display(oct18_s1)

print("\n📅 26th October Slot S2 Results:")
oct26_s2 = calculate_loyalty_points(df, pd.to_datetime('2022-10-26').date(), 'S2')
display(oct26_s2)
```

**2. Calculate overall loyalty points earned and rank players on the basis of loyalty points in the month of October.
     In case of tie, number of games played should be taken as the next criteria for ranking.**

```python
def calculate_october_loyalty(df):
    df['Date'] = pd.to_datetime(df['Date'])
    october_data = df[df['Date'].dt.month == 10]
    monthly = october_data.groupby('User Id').agg(
        Total_Deposits=('Amount', lambda x: x[x > 0].sum()),
        Total_Withdrawals=('Amount', lambda x: abs(x[x < 0].sum())),
        Deposit_Count=('Amount', lambda x: (x > 0).sum()),
        Withdrawal_Count=('Amount', lambda x: (x < 0).sum()),
        Total_Games=('Games_Played', 'sum')
    ).reset_index()
    monthly['Loyalty_Points'] = (
        0.01 * monthly['Total_Deposits'] +
        0.005 * monthly['Total_Withdrawals'] +
        0.001 * np.maximum(monthly['Deposit_Count'] - monthly['Withdrawal_Count'], 0) +
        0.2 * monthly['Total_Games']
    )
    monthly = monthly.sort_values(by=['Loyalty_Points', 'Total_Games'], ascending=[False, False])
    monthly['Rank'] = range(1, len(monthly) + 1)
    return monthly
october_rankings = calculate_october_loyalty(df)

print("🏆 Top 50 Players for October (based on loyalty points):")
display(october_rankings.head(50))
```

Part B - How much bonus should be allocated to leaderboard players?
After calculating the loyalty points for the whole month find out which 50 players are at the top of the leaderboard. The company has allocated a pool of Rs 50000 to be given away as bonus money to the loyal players.Now the company needs to determine how much bonus money should be given to the players.Should they base it on the amount of loyalty points? Should it be based on number of games? Or something else?That’s for you to figure out.
Suggest a suitable way to divide the allocated money keeping in mind the following points:1. Only top 50 ranked players are awarded bonus.

```python
BONUS_POOL = 50000
top_50 = october_rankings.head(50).copy()

top_50['Norm_Loyalty'] = top_50['Loyalty_Points'] / top_50['Loyalty_Points'].max()
top_50['Norm_Games'] = top_50['Total_Games'] / top_50['Total_Games'].max()

top_50['Final_Score'] = 0.7 * top_50['Norm_Loyalty'] + 0.3 * top_50['Norm_Games']

top_50['Allocated_Bonus'] = (top_50['Final_Score'] / top_50['Final_Score'].sum()) * BONUS_POOL
top_50['Allocated_Bonus'] = top_50['Allocated_Bonus'].round(2)

print("🎉 Hybrid Bonus Allocation (Top 50 Players):")
display(top_50[['User Id', 'Rank', 'Loyalty_Points', 'Total_Games', 'Final_Score', 'Allocated_Bonus']].head(50))

print(f"✅ Total Bonus Distributed: ₹{top_50['Allocated_Bonus'].sum():.2f}")
```

sol:Final Score = 70% Loyalty Points + 30% Normalized Games Played
This balances loyalty and platform engagement:Rewards high loyalty scorers,But also gives weight to players who are active.
Insights from this method:
Players with both high monetary value and high engagement get more bonus.
Less active users with very high deposits are still rewarded.
Very active users (even with moderate deposits) are not ignored.

Would you say the loyalty point formula is fair or unfair?

Can you suggest any way to make the loyalty point formula more robust?

The current loyalty point formula is relatively fair as it transparently rewards users for both monetary activity (deposits and withdrawals) and gameplay. It encourages user participation by assigning weight to deposits and number of games played, which aligns well with business objectives like user engagement and revenue generation. However, the formula is not without limitations. The heavier weight on deposits may unintentionally encourage users to deposit repeatedly without genuine engagement, while the minimal weight on games played undervalues consistent player activity. Additionally, there is no penalty for users who withdraw more frequently or are inactive for extended periods. To make the system more robust, the formula could be enhanced by balancing the financial weights, adding a bonus for games won to reward skill, applying multipliers for consistent activity, and introducing decay for inactivity. These improvements would create a more balanced and fair system that encourages not just financial transactions but also sustained and meaningful platform engagement.
