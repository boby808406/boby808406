**When Are Bookings Typically Made?**

```python
# Analyze Prime time
hotel['total_stay'] = hotel['stays_in_weekend_nights'] + hotel['stays_in_week_nights']
plt.figure(figsize=(12, 6))
sns.histplot(hotel['lead_time'], bins=50, kde=True, color='blue')
plt.title('Distribution of Prime Time (Days)', fontsize=20)
plt.xlabel('Prime Time (Days)', fontsize=16)
plt.ylabel('Frequency', fontsize=16)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Analyze bookings by month
hotel['reservation_status_date'] = pd.to_datetime(hotel['reservation_status_date'])
hotel['reservation_month'] = hotel['reservation_status_date'].dt.month_name()
monthly_reservations = hotel['reservation_month'].value_counts().reindex( ['January', 'February', 'March', 'April', 'May', 'June','July', 'August', 'September', 'October', 'November', 'December'])
plt.figure(figsize=(12, 6))
monthly_reservations.plot(kind='bar', color='red', alpha=0.8)
plt.title('Number of Bookings by Reservation Month', fontsize=16)
plt.xlabel('Month', fontsize=14)
plt.ylabel('Number of Bookings', fontsize=14)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

**Pricing Strategies**

```python
# Correlation between length of stay and ADR
hotel = hotel[(hotel['adr'] > 0) & (hotel['total_stay'] > 0)]
correlation = hotel[['total_stay', 'adr']].corr().iloc[0, 1]
print(f"Correlation between length of stay and ADR: {correlation:.2f}")
plt.figure(figsize=(12, 6))
sns.scatterplot(data=hotel, x='total_stay', y='adr', alpha=0.6, color='green')
plt.title('Scatterplot of Length of Stay vs ADR', fontsize=20)
plt.xlabel('Total Length of Stay (Nights)', fontsize=16)
plt.ylabel('ADR (Average Daily Rate)', fontsize=16)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Average ADR by stay length
stay_adr = hotel.groupby('total_stay')['adr'].mean().reset_index()
plt.figure(figsize=(12, 6))
sns.lineplot(data=stay_adr, x='total_stay', y='adr', marker='o', color='red')
plt.title('Average ADR by Length of Stay',fontsize=20)
plt.xlabel('Total Length of Stay (Nights)',fontsize=16)
plt.ylabel('Average ADR',fontsize=16)
plt.grid(axis='y', linestyle='--')
plt.show()
```

**Factors Leading to Special Requests**

```python
# Distribution of special requests
plt.figure(figsize=(10, 6))
sns.countplot(data=hotel, x='total_of_special_requests', palette='viridis')
plt.title('Distribution of Total Special Requests',fontsize=20)
plt.xlabel('Number of Special Requests',fontsize=16)
plt.ylabel('Frequency', fontsize=16)
plt.grid(axis='y', linestyle='--')
plt.show()
```

```python
# Special requests by customer type
plt.figure(figsize=(10, 6))
sns.boxplot(data=hotel, x='customer_type', y='total_of_special_requests', palette='Set3')
plt.title('Special Requests by Customer Type',fontsize=20)
plt.xlabel('Customer Type',fontsize=16)
plt.ylabel('Total Special Requests',fontsize=16)
plt.grid(axis='y', linestyle='--')
plt.show()
```

**Demographic Factors Influencing Booking Patterns**

```python
# Top countries for bookings
top_countries = hotel['country'].value_counts().head(20)
plt.figure(figsize=(10, 6))
sns.barplot(x=top_countries.index, y=top_countries.values, palette='viridis')
plt.title('Top 20 Countries by Number of Bookings', fontsize=20)
plt.xlabel('Country', fontsize=16)
plt.ylabel('Number of Bookings', fontsize=16)
plt.xticks(rotation=90)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Family size influence on booking patterns
hotel['family_size'] = hotel['adults'] + hotel['children'] + hotel['babies']
plt.figure(figsize=(10, 6))
sns.countplot(data=hotel, x='family_size', palette='Set2')
plt.title('Booking Patterns by Family Size', fontsize=20)
plt.xlabel('Family Size (Number of People)', fontsize=16)
plt.ylabel('Number of Bookings', fontsize=16)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Market segment analysis
plt.figure(figsize=(10, 6))
sns.countplot(data=hotel, x='market_segment', order=hotel['market_segment'].value_counts().index, palette='Set3')
plt.title('Booking Patterns by Market Segment', fontsize=20)
plt.xlabel('Market Segment', fontsize=16)
plt.ylabel('Number of Bookings', fontsize=16)
plt.xticks(rotation=90)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

**Percentage of Repeated Guests**

```python
repeated_guests = hotel['is_repeated_guest'].value_counts(normalize=True) * 100
plt.figure(figsize=(10, 6))
plt.pie(repeated_guests, labels=['New Guest', 'Repeated Guest'], autopct='%1.1f%%', startangle=90, colors=['skyblue', 'red'])
plt.title('Percentage of Repeated Guests', fontsize=16)
plt.show()
```
